```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("alexa_reviews_dataset.tsv", sep = "\t")
df
```

```python
dataset = df[["verified_reviews", "rating"]]
dataset.columns =  ["Review", "Sentiment"]
dataset.head()
```

```python
def compute_sentiments(labels):
    sentiments = []

    for label in labels:
        if label > 3.0:
            sentiment = 1
        elif label <= 3.0:
            sentiment = 0

        sentiments.append(sentiment)
    return sentiments
```

```python
dataset["Sentiment"] = compute_sentiments(dataset.Sentiment)
```
